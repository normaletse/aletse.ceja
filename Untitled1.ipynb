```python
import pandas as pd
from scipy import stats


# 1. Asegurar que user_score sea numérico
df["user_score"] = pd.to_numeric(df["user_score"], errors="coerce")

# 2. Filtrar las calificaciones de usuarios para XOne y PC
xone_scores = df[df["platform"] == "XOne"]["user_score"].dropna()
pc_scores = df[df["platform"] == "PC"]["user_score"].dropna()

print("Cantidad de calificaciones XOne:", len(xone_scores))
print("Cantidad de calificaciones PC:", len(pc_scores))

# 3. Prueba t de dos muestras independientes
t_stat, p_value = stats.ttest_ind(xone_scores, pc_scores, equal_var=False)

print("\n=== RESULTADOS DE LA PRUEBA T ===")
print("t-statistic:", t_stat)
print("p-value:", p_value)

# 4. Interpretación del resultado
alpha = 0.05
print("\n=== CONCLUSIÓN ===")
if p_value < alpha:
    print("Rechazamos H0 → Las calificaciones promedio de XOne y PC NO son iguales.")
else:
    print("No se puede rechazar H0 → No hay evidencia suficiente para decir que las calificaciones promedio difieren.")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# 1. Establecer el valor de alfa
alpha = 0.05   # umbral alfa

# 2. Filtrar solo los géneros deseados
df_gen = df[df["genre"].isin(["Action", "Sports"])].copy()

# 3. Convertir user_score a numérico
df_gen["user_score"] = pd.to_numeric(df_gen["user_score"], errors="coerce")

# Eliminar valores vacíos
df_gen = df_gen.dropna(subset=["user_score"])

# 4. Separar los grupos
action_scores = df_gen[df_gen["genre"] == "Action"]["user_score"]
sports_scores = df_gen[df_gen["genre"] == "Sports"]["user_score"]

print("Cantidad de valores Acción:", len(action_scores))
print("Cantidad de valores Deportes:", len(sports_scores))

# 5. Prueba de hipótesis (t-test)
t_stat, p_value = ttest_ind(action_scores, sports_scores, equal_var=False)
print("\n----- RESULTADOS DEL TEST -----")
print("Estadístico t:", t_stat)
print("p-value:", p_value)

if p_value < alpha:
    print(f"\nComo p < {alpha}, rechazamos H0.")
    print("Conclusión: Las calificaciones promedio de Acción y Deportes son diferentes.")
else:
    print(f"\nComo p >= {alpha}, NO rechazamos H0.")
    print("Conclusión: No hay evidencia suficiente para decir que los promedios sean diferentes.")

# 6. Gráfico comparativo
plt.figure(figsize=(8,6))
plt.boxplot(
    [action_scores, sports_scores],
    labels=["Action", "Sports"]
)

plt.title("Comparación de Calificaciones de Usuarios\nAcción vs Deportes")
plt.ylabel("User Score")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()







```

¿Como formulaste las hipotesis nula y alternativa?
las calificaciones para accion y deportes es la misma, no existe referencia real en las medidas
las calificaciones promedio son diferentes

Para probar la hipotesis el criterio que se uso fue, probar las hipótesis se utilizó el valor p (p-value) obtenido mediante una prueba t de dos muestras independientes
